```python
for col in df.select_dtypes(include="object").columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.replace(r"\s+", " ", regex=True)
        .str.lower()
        .replace({"nan": pd.NA})
    )

    # ---------------------------------------------
# 2. NORMALIZED YES/NO MAPPING DICTIONARY
# ---------------------------------------------
yes_no_map = {
    "yes": 1,
    "no": 0,
    "dependss": 0,

    # missing / garbage
    "don't know": pd.NA,
    "dont know": pd.NA,
    "do not know": pd.NA,
}

for col in yes_no_cols:
    if col in df.columns:
        df[col] = (
            df[col]
            .replace(yes_no_map)                 # map yes/no
            .where(lambda s: s.isin([0, 1]), pd.NA)  # drop garbage safely
            .astype("Float64")                  # nullable binary
        )
```

# 6. Likert columns

```python
for col in ["att_free_safe_access", "att_child_marriage", "att_community_engage_gbv", "att_services_welcoming"]:
    print(f"\n{col}:")
    print(df[col].value_counts(dropna=False))
```

```python
# 1. Likert variables
likert_vars = ["att_free_safe_access", "att_community_engage_gbv"]

likert_map = {
    "strongly agree": 1,
    "agree": 1,
    "disagree": 0,
    "strongly disagree": 0
}

for col in likert_vars:
    df[col] = df[col].map(likert_map)

# 2. Child marriage (custom scale)

df["att_child_marriage"] = df["att_child_marriage"].map({
    "no": 1,
    "yes": 0,
    "depends": 0
})

# 3. Services welcoming (binary)
df["att_services_welcoming"] = df["att_services_welcoming"].map({
    "yes": 1,
    "no": 0,
    "don't know": None
})
```

```python
df["n_children"] = df["n_children"].astype(str).str.strip()

df["n_children"] = (
    df["n_children"]
    .replace(["", "none", "NaN", "nan"], "0")
    .replace(["1", "one"], "1")
    .replace(["more than 1", "more_than_one", "More than 1"], "2")
)

df["n_children"] = pd.to_numeric(
    df["n_children"],
    errors="coerce"
)

# THIS is what you are missing
df["n_children"] = df["n_children"].fillna(0).astype(int)

df["n_children"].value_counts(dropna=False)
```

```python
numeric_cols = [
    'age_first_child',
    'interval_first_second_child',
    'age_years',
    'n_children'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.loc[
    (df["n_children"] == 0) &
    (df["age_first_child"] == 0),
    "age_first_child"
] = pd.NA


df.loc[
    df["age_first_child"].notna() & (df["n_children"] == 0),
    "n_children"
] = 1
df.loc[df["interval_first_second_child"].notna(), "n_children"] = 2

df["n_children"] = df["n_children"].astype("Int64")
```

# 4. Save the open ended text in different dataframe for later classification

```python
META_COLS = ["id", "district", "community", "participant_profile"]
TEXT_COLS_Q1 = [
    "att_free_safe_access_comment",
    "att_child_marriage_comment",
    "att_community_engage_gbv_comment",
    "talked_to_parents_comment",
    "support_improvement",
    "savings_participation_comment",
    "changes_seen",
    "recommendations",
]
# Step 2: reshape to long format
df1_open = (
    df[META_COLS + TEXT_COLS_Q1]
    .melt(
        id_vars=META_COLS,
        value_vars=TEXT_COLS_Q1,
        var_name="text_type",
        value_name="text"
    )
    .dropna(subset=["text"])
)

df1_open["source"] = "q1"
# Save text dataset
df1_open.to_parquet(
    "/Users/anamargarida/Local/Coding/Tsogolo Tsicana/data_clean/open_text/q1_open_ended.parquet",
    index=False
)

df = df.drop(columns=TEXT_COLS_Q1)
df
```
